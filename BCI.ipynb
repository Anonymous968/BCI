```python
import scipy.io as spio
import numpy as np
import tensorflow as tf
import keras
from keras.models import Sequential
from keras.layers import Dense,Dropout
```

```python
mat1 = spio.loadmat(r'C:\Chanakya\Textbooks\Semester-III\Project Seminar\BCI_left_filter.m', squeeze_me=True)
mat2 = spio.loadmat(r'C:\Chanakya\Textbooks\Semester-III\Project Seminar\BCI_right_filter.m', squeeze_me=True)
```

```python
left_filter=mat1['left_filter_comb']
right_filter=mat2['right_filter_comb']
```

```python
left_filter.shape
right_filter.shape
```

```python
left_filter[1,:,:].shape
```

```python
x1=np.transpose(left_filter[1,:,:])
print(x1.shape)
x2=np.transpose(right_filter[1,:,:])
print(x2.shape)
```

```python
combined=np.concatenate((x1,x2))
print(combined.shape)
x_train=combined
```

```python
p1=np.ones((18,1))
print(p1.shape)
p2=np.zeros((18,1))
print(p2.shape)
combined_p=np.concatenate((p1,p2))
print(combined_p.shape)
y_train=combined_p
print(y_train)
```

```python
model=Sequential()

model.add(Dense(48, input_dim=12, activation='tanh'))
model.add(Dropout(0.4))
model.add(Dense(48, activation='tanh'))
model.add(Dropout(0.4))
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='rmsprop', metrics=['accuracy'])

model.fit(x_train, y_train, epochs=400, batch_size=36)
```
